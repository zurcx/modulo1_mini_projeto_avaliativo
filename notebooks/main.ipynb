## 🔍 1. Diagnóstico — Entendendo os Dados Antes de Limpar

**Nunca saia limpando sem entender o problema primeiro!**

Um bom diagnóstico responde:
- Quantas linhas e colunas temos?
- Quais colunas têm valores faltando? Quanto?
- Há registros duplicados?
- Os tipos de dados fazem sentido?

Vamos criar uma função de diagnóstico que usaremos sempre:


```python
def diagnostico(df, nome="DataFrame"):
    """
    Mostra um relatório completo de qualidade dos dados.
    Use sempre antes de começar a limpeza!
    """
    print("=" * 55)
    print(f"  📊 DIAGNÓSTICO: {nome}")
    print("=" * 55)
    print(f"  Linhas:           {df.shape[0]:,}")
    print(f"  Colunas:          {df.shape[1]}")
    print(f"  Linhas duplicadas:{df.duplicated().sum():,}")
    print()

    nulos = df.isnull().sum()
    pct   = (nulos / len(df) * 100).round(1)

    print("  Coluna           | Tipo       | Nulos | % Nulos")
    print("  " + "-"*50)
    for col in df.columns:
        tipo = str(df[col].dtype)
        print(f"  {col:<18}| {tipo:<10} | {nulos[col]:<5} | {pct[col]}%")
    print("=" * 55)

# Rodando o diagnóstico no nosso dataset
diagnostico(df, "Pedidos E-commerce")
```

## ❌ 2. Valores Ausentes

**O que são valores ausentes?**

São células que não têm informação — campo em branco, dado que não foi preenchido ou veio corrompido.

No Python/Pandas, um valor ausente pode aparecer de várias formas:

| Como aparece | O que significa |
|---|---|
| `NaN` | Not a Number — o padrão do Pandas para "sem valor" |
| `None` | O Python nativo para "nulo" |
| `''` (vazio) | String vazia — **não é NaN!** O Pandas não identifica automaticamente |
| `'NULL'`, `'N/A'` | Strings que representam nulo — também não são NaN! |

> ⚠️ **Cuidado:** `isnull()` só detecta NaN e None. Strings vazias e 'NULL' precisam ser tratadas separadamente!


```python
# ── Demonstração com exemplo simples ────────────────────
df_ex = pd.DataFrame({
    'nome':    ['Ana', None, 'Carlos', ''],
    'idade':   [25, 30, np.nan, 28],
    'status':  ['pago', 'NULL', 'pago', 'N/A'],
})

print("Nosso DataFrame de exemplo:")
print(df_ex)
print()

# isnull() mostra onde há NaN e None
print("Onde há NaN/None? (True = tem problema)")
print(df_ex.isnull())
```

```python
# Contando nulos por coluna
print("Quantidade de nulos por coluna:")
print(df_ex.isnull().sum())
print()

# Repare: '' e 'NULL' e 'N/A' NÃO foram detectados!
# Precisamos tratar esses casos manualmente (veja abaixo)
```

```python
# ── Detectando strings que representam nulo ─────────────
# Passo 1: Converter strings 'NULL', 'N/A', '' para np.nan
df_limpo = df_ex.replace({'NULL': np.nan, 'N/A': np.nan, '': np.nan})

print("Depois de converter strings vazias/NULL/N/A:")
print(df_limpo)
print()
print("Agora o isnull() detecta tudo corretamente:")
print(df_limpo.isnull().sum())
```

```python
# ── Detectando strings vazias com espaço ────────────────
# '   ' (só espaços) também não é detectado pelo isnull()!

df_espaco = pd.DataFrame({'valor': ['ok', '   ', 'outro', '']})
print("Strings com só espaço:")
print(df_espaco)
print()
print("isnull() não detecta:  ", df_espaco['valor'].isnull().sum(), "nulos")

# Solução: verificar se str.strip() resulta em string vazia
eh_vazio = df_espaco['valor'].str.strip() == ''
print("Após str.strip() == '': ", eh_vazio.sum(), "vazios")
```

### 2.1 Como Tratar Valores Ausentes?

Existem **3 estratégias**. A escolha depende do contexto:

| Estratégia | Quando usar |
|---|---|
| **Remover a linha** | Quando a coluna é crítica e não há como recuperar o dado |
| **Preencher** | Quando não queremos perder a linha — usamos média, mediana, moda ou um valor fixo |
| **Interpolar** | Para séries temporais, quando os dados têm sequência temporal |


```python
# ── ESTRATÉGIA 1: Remover linhas com nulo ───────────────

df_num = pd.DataFrame({
    'id':      [1, 2, 3, 4, 5],
    'produto': ['A', 'B', None, 'D', 'E'],
    'preco':   [10.0, np.nan, 30.0, 40.0, np.nan],
})

print("Original:", df_num.shape)
print(df_num)
print()

# Remover qualquer linha que tenha pelo menos 1 nulo
df_drop = df_num.dropna()
print("Após dropna() — remove linha com qualquer nulo:")
print(df_drop)
print()

# Remover apenas onde uma coluna específica é nula
# (mais preciso — só remove se 'produto' for nulo)
df_drop2 = df_num.dropna(subset=['produto'])
print("Após dropna(subset=['produto']):")
print(df_drop2)
```

```python
# ── ESTRATÉGIA 2: Preencher nulos (fillna) ──────────────

df_fill = pd.DataFrame({
    'nome':    ['Ana', None, 'Carlos', None, 'Eduardo'],
    'salario': [3000.0, 5000.0, np.nan, 4000.0, np.nan],
    'cidade':  ['SP', 'RJ', None, 'SP', 'MG'],
})

# Preencher com valor fixo (para texto)
df_fill['nome']   = df_fill['nome'].fillna('Desconhecido')
df_fill['cidade'] = df_fill['cidade'].fillna('Não informado')

# Preencher com mediana (para números — mais seguro que média quando há outliers)
mediana_sal = df_fill['salario'].median()
print(f"Mediana do salário: R$ {mediana_sal:.2f}")
df_fill['salario'] = df_fill['salario'].fillna(mediana_sal)

print()
print("Resultado após fillna:")
print(df_fill)
```

```python
# ── Média vs Mediana: quando usar cada uma? ─────────────

# Exemplo: um dataset com um salário absurdo (outlier)
salarios = pd.Series([3000, 3500, 4000, 4200, 4500, 100000])

print(f"Salários: {salarios.values}")
print(f"Média:    R$ {salarios.mean():,.2f}  ← distorcida pelo 100.000!")
print(f"Mediana:  R$ {salarios.median():,.2f}  ← representa melhor a maioria")
print()
print("👉 Regra: se há outliers na coluna → use MEDIANA")
print("   Se os dados são bem distribuídos  → pode usar MÉDIA")
```

```python
# ── ESTRATÉGIA 3: Interpolar (para séries temporais) ────

# Quando temos dados em sequência de tempo, faz sentido
# estimar o valor pelo que veio antes e depois

temperatura = pd.DataFrame({
    'hora':  ['08:00','09:00','10:00','11:00','12:00'],
    'graus': [22.1, None, None, 24.8, 25.3]
})

print("Leituras do sensor com falhas:")
print(temperatura)
print()

# Interpolação linear: estima os valores faltando
temperatura['graus_ok'] = temperatura['graus'].interpolate()
print("Após interpolação (estimativa entre os pontos):")
print(temperatura)
```

```python
# ── Aplicando no nosso dataset real ─────────────────────

print("Diagnóstico ANTES da limpeza de nulos:")
print(df.isnull().sum())
print()

df_limpo = df.copy()  # SEMPRE trabalhe com uma cópia!

# Passo 1: converter strings que representam nulo
df_limpo = df_limpo.replace({'NULL': np.nan, 'N/A': np.nan, '': np.nan})

# Passo 2: verificar o que melhorou
print("Depois de converter NULL/N/A para np.nan:")
print(df_limpo.isnull().sum())
```

## 🔄 3. Duplicatas

**O que são duplicatas?**

São linhas repetidas no dataset — o mesmo registro aparece mais de uma vez. Causas comuns:
- Botão "Salvar" clicado duas vezes
- Falha na integração entre sistemas
- Exportação duplicada de dados

**Por que são um problema?** Uma duplicata de um pedido de R$ 1.000 vai fazer seu total aparecer como R$ 2.000!


```python
# ── Detectando duplicatas ────────────────────────────────

df_dup_ex = pd.DataFrame({
    'id':     [1, 2, 3, 2, 4, 3],
    'nome':   ['Ana','Bruno','Carla','Bruno','Diana','Carla'],
    'valor':  [100, 200, 300, 200, 400, 300]
})

print("Dataset com duplicatas:")
print(df_dup_ex)
print()

# Verificar linhas completamente iguais
print(f"Linhas completamente iguais: {df_dup_ex.duplicated().sum()}")
print()

# Verificar por coluna específica (ID duplicado)
print(f"IDs duplicados: {df_dup_ex.duplicated(subset=['id']).sum()}")
print()

# Ver quais linhas estão duplicadas
print("Linhas com ID duplicado (todas as cópias):")
print(df_dup_ex[df_dup_ex.duplicated(subset=['id'], keep=False)])
```

```python
# ── O parâmetro keep: qual cópia manter? ────────────────

# keep='first' → mantém a PRIMEIRA ocorrência (mais comum)
df1 = df_dup_ex.drop_duplicates(subset=['id'], keep='first')
print("keep='first' (manter a primeira aparição):")
print(df1)
print()

# keep='last' → mantém a ÚLTIMA ocorrência
# Útil quando os dados mais recentes são mais confiáveis
df2 = df_dup_ex.drop_duplicates(subset=['id'], keep='last')
print("keep='last' (manter a última — mais recente):")
print(df2)
```

```python
# ── Aplicando no dataset real ────────────────────────────

print(f"ANTES: {len(df_limpo)} linhas")
print(f"IDs de pedido duplicados: {df_limpo.duplicated(subset=['id_pedido']).sum()}")
print()

# Removendo duplicatas pelo id_pedido, mantendo o primeiro
df_limpo = df_limpo.drop_duplicates(subset=['id_pedido'], keep='first')

# ⚠️ Sempre reindexe após remover linhas!
# Sem isso, os índices ficam com "buracos" (0,1,5,8,...)
df_limpo = df_limpo.reset_index(drop=True)

print(f"DEPOIS: {len(df_limpo)} linhas")
print(f"Duplicatas restantes: {df_limpo.duplicated(subset=['id_pedido']).sum()}")
```

## ✅ Resumo do Dia 1

| Função | O que faz |
|---|---|

| `df.replace({'NULL': np.nan})` | Converte strings para NaN |
| `df.dropna()` | Remove linhas com nulo |
| `df.dropna(subset=['col'])` | Remove apenas se coluna específica for nula |
| `df['col'].fillna(valor)` | Preenche nulos com um valor |

| `df['col'].interpolate()` | Interpola valores faltando (séries temporais) |
| `df.duplicated().sum()` | Conta linhas duplicadas |
| `df.drop_duplicates(subset=['id'])` | Remove duplicatas por coluna |
| `df.reset_index(drop=True)` | Reindexar após remover linhas |
